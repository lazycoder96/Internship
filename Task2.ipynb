# Loading CSV file and printing Data

```python
import pandas as pd
dataframe = pd.read_csv('C:\\Users\\adars\\01.Data Cleaning and Preprocessing.csv')
print(dataframe)
```

## FILTERING DATA

# Filtering Data where "Y-Kappa" is less than 20.

```python
dataframe1 = dataframe.loc[dataframe["Y-Kappa"]<20]
print(dataframe1)
```

# Filtering Data where "Y-Kappa" is less than 25 and "Chiprate" is less than 20 and  "BF-CMratio" is less than 90

```python
dataframe2 = dataframe.loc[(dataframe["Y-Kappa"]<25 ) & (dataframe["ChipRate"]<20) & (dataframe["BF-CMratio"]<90)]
print(dataframe2)
```

## HANDLING NULL VALUES

# checking null values from each column

```python
dataframe3 = dataframe.isnull().sum()
print(dataframe3)
```

# Counting total number of null values

```python
dataframe0 = dataframe.isnull().sum().sum()
print(dataframe0)
```

# Removing Null Values

```python
dataframe4 = dataframe.dropna()
print(dataframe4)
```

# Filling Null Values with 0

```python
dataframe5 = dataframe.fillna(value = 0)
print(dataframe5)
```

# Dropping Duplicate values that are null

```python
dataframe6 = dataframe.drop_duplicates()
print(dataframe6)
```

# Calculating Summary Statistics

```python
dataframe7 = dataframe.describe()
print(dataframe7)
```

```python
dataframe8 = dataframe.describe(include = ['object'])
print(dataframe8)
```
