## Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Loading DataSet

```python
df = pd.read_csv("C:\\Users\\adars\\Downloads\\USvideos.csv")
df.head()
```

```python
#Replacing infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
#Handling Null Values
df.dropna(inplace=True)
```

## Visualization to understand the distrubution

```python
# Histogram of Views
plt.figure(figsize=(6, 3))
sns.histplot(df['views'], bins=50, kde=True, color='blue')
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of 'likes' by 'category_id'

plt.figure(figsize=(6, 3))
sns.boxplot(x='category_id', y='likes', data=df)
plt.title('Boxplot of Likes by Category')
plt.xlabel('Category ID')
plt.ylabel('Likes')
plt.xticks(rotation=90)
plt.show()
```

```python
# Boxplot of 'dislikes' by 'category_id'
plt.figure(figsize=(6, 3))
sns.boxplot(x='category_id', y='dislikes', data=df)
plt.title('Boxplot of Dislikes by Category')
plt.xlabel('Category ID')
plt.ylabel('Dislikes')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter Plot of likes vs dislikes
plt.figure(figsize=(6, 3))
sns.scatterplot(x='likes', y='dislikes', data=df)
plt.title('Scatter Plot of Likes vs Dislikes')
plt.xlabel('Likes')
plt.ylabel('Dislikes')
plt.show
```

```python
plt.tight_layout()
```

```python
#Correlatin Heatmap of all variables
plt.figure(figsize=(8, 8))
sns.heatmap(df[['views', 'likes', 'dislikes', 'comment_count']].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
