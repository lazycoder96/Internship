## Creating a Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
dataframe = pd.read_csv("C:\\intern\\tipscsv1.csv")
plt.bar(dataframe['day'],dataframe['tip'])
plt.title("Bar Chart")
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

## Creating a Line Chart

```python
import pandas as pd 
import matplotlib.pyplot as plt
df2 = pd.read_csv("C:\\intern\\tipscsv1.csv")
plt.plot(df2['tip'],label='Tip')
plt.plot(df2['size'],label='Days')
plt.title('Line- Chart')
#Adding Labels
plt.xlabel('Size')
plt.ylabel('tip')
#Adding Legends
plt.legend()
plt.show()
```
